```python
# Lee el archivo APPL.csv
df = pd.read_csv(r"DATOS/TECNO/AAPL.csv")
```

```python
df.head()
```

```python

# convierte la columna de fechas a formato de fecha
df['Date'] = pd.to_datetime(df['Date'])

# obtiene el nombre del día de la semana para cada fecha
df['Days'] = df['Date'].dt.day_name()
```

```python
# Visualizar las primeras filas del dataframe
df.head()
```

```python
# Obtener información general del dataframe
df.info()
```

```python
# Obtener estadísticas descriptivas del dataframe
df.describe()
```

```python
# Graficar un histograma de la variable 'Close'
sns.histplot(data=df, x='Close')
plt.show()
```

```python
# Graficar un diagrama de caja de la variable 'Volume'
sns.boxplot(data=df, y='Volume')
plt.show()
```

```python
# Graficar la relación entre las variables 'Close' y 'Volume'
sns.scatterplot(data=df, x='Close', y='Volume')
plt.show()
```

```python
# Graficar la evolución de la variable 'Close' en el tiempo
sns.lineplot(data=df, x='Date', y='Close')
plt.show()
```

```python
# Análisis exploratorio de la variable "Open"
sns.histplot(data=df, x="Open", kde=True)
plt.title("Distribución de los valores de apertura (Open)")
plt.xlabel("Valor de apertura")
plt.ylabel("Frecuencia")
plt.show()

```

```python
# Análisis exploratorio de la variable "Volume"
sns.boxplot(data=df, x="Days", y="Volume")
plt.title("Distribución del volumen por día de la semana")
plt.xlabel("Día de la semana")
plt.ylabel("Volumen")
plt.show()
```

```python

# Calcular el promedio del volumen para cada día de la semana
avg_volume_by_day = df.groupby("Days")["Volume"].mean()

# Crear un gráfico de barras con los promedios del volumen por día de la semana
sns.barplot(x=avg_volume_by_day.index, y=avg_volume_by_day.values)
plt.title("Volumen promedio por día de la semana")
plt.xlabel("Día de la semana")
plt.ylabel("Volumen promedio")
plt.show()

```

```python


# Identificación de valores atípicos utilizando el método del rango intercuartílico
Q1 = df["Open"].quantile(0.25)
Q3 = df["Open"].quantile(0.75)
IQR = Q3 - Q1
df[(df["Open"] < Q1 - 1.5*IQR) | (df["Open"] > Q3 + 1.5*IQR)]
```

```python
# Scatterplot de las variables "Open" y "Close"
sns.scatterplot(data=df, x="Open", y="Close")
plt.show()
```

```python
df['retornos_diarios'] = (df['Adj Close'] / df['Adj Close'].shift(1)) - 1
```

```python
df.head()
```

```python
#calculo de la tasa de retorno promedio diaria
daily_return_mean = df['retornos_diarios'].mean()
daily_return_mean
```

```python
#calculo de la tasa de riesgo de inversion
daily_return_std = df['retornos_diarios'].std()
daily_return_std
```
