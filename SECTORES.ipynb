```python
df1['sector'] = 'indice'
df2['sector'] = 'salud'
df3['sector'] = 'comunicacion'
df4['sector'] = 'cons. basicos'
df5['sector'] = 'cons discrecional'
df6['sector'] = 'energia'
df7['sector'] = 'finanzas'
df8['sector'] = 'industrial'
df9['sector'] = 'inmobiliario'
df10['sector'] = 'materiales'
df11['sector'] = 'ss. publicos'
df12['sector'] = 'tecnologia'
```

```python
df1 = df1.drop('daily_returns', axis=1)
```

```python
# Concatenar dataframes por fila
df = pd.concat([df1, df2, df3, df4, df5, df6, df7, df8, df9, df10, df11, df12], axis=0)
```

```python
df.tail()
```

```python
#calcular el rendimiento diario y agregar la columna
df['rendimiento_diario'] = df['Adj Close'].pct_change()
```

```python
df.tail()
```

```python
#separar la fecha en dia meses años
df = df.assign(year=pd.to_datetime(df['Date']).dt.year,
               month=pd.to_datetime(df['Date']).dt.month,
               day=pd.to_datetime(df['Date']).dt.day)
```

```python
df.tail()
```

```python
# Obtener la cantidad de filas y columnas del DataFrame
num_filas = df.shape[0]
num_columnas = df.shape[1]

# Obtener la cantidad de valores nulos en cada columna
valores_nulos = df.isnull().sum()
```

```python
# Calcular la desviación estándar de los rendimientos diarios por sector
volatility_by_sector = df.groupby('sector')['rendimiento_diario'].std()

# Agregar una columna al DataFrame con la volatilidad por sector
df['volatilidad'] = df['sector'].map(volatility_by_sector)
```

```python
df.head()
```

```python


# Agrupar por sector y calcular el promedio del riesgo
riesgo_promedio = df.groupby('sector')['volatilidad'].mean()

# Crear el gráfico de barras
fig, ax = plt.subplots(figsize=(12, 6))
sns.barplot(x=riesgo_promedio.index, y=riesgo_promedio.values, ax=ax)
ax.set_xlabel('Sector')
ax.set_ylabel('volatilidad')
ax.set_title('Riesgo de inversión por sector')

# Rotar las etiquetas del eje x
plt.xticks(rotation=90)

# Ajustar las leyendas verticales
plt.tight_layout()
```
